```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
```

```python
import os
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
train_data = pd.read_csv("/kaggle/input/c/titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/c/titanic/test.csv")
test_data.head()
```

```python
#Saber que porcentaje de todas las mujeres a bordo sobrevivieron.
women = train_data.loc[train_data.Sex == 'female']["Survived"]
rate_women = sum(women)/len(women)

print("% de mujeres que sobrevivieron:", rate_women)
```

```python
#Saber que porcentaje de todas los hombres a bordo sobrevivieron.
men = train_data.loc[train_data.Sex == 'male']["Survived"]
rate_men = sum(men)/len(men)

print("% de hombres que sobrevivieron:", rate_men)
```

```python
# Importar bibliotecas necesarias para la visualizacion de datos 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#Trazar un gráfico de barras de supervivencia por género
sns.barplot(x='Sex', y='Survived', data=train_data)
```

```python
#Gráfico de pastel para la distribución de género:
plt.figure(figsize=(6,6))
labels = ['Hombres', 'Mujeres']
sizes = [train_data['Sex'].value_counts()[0], train_data['Sex'].value_counts()[1]]
colors = ['skyblue', 'pink']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Distribución de género')
plt.show()
```

```python
# Crear un histograma de la variable 'Edad'
import matplotlib.pyplot as plt
plt.hist(train_data['Age'].dropna(), bins=20)
plt.show()
```

```python
#Gráfico de barras para la distribución de edades
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
plt.figure(figsize=(12,6))
sns.countplot(x='Age', data=train_data, color='skyblue')
plt.title('Distribución de edades')
plt.xlabel('Edades')
plt.ylabel('Cantidad')
plt.show()
```

```python
# Crea un gráfico de dispersión de 'Edad' vs. 'tarifa'
plt.scatter(train_data['Age'], train_data['Fare'])
plt.xlabel('Age')
plt.ylabel('Fare')
plt.show()
```

```python
#Gráfico de cajas para mostrar la relación entre la tarifa y la clase de pasajero:
plt.figure(figsize=(10,6))
sns.boxplot(x='Pclass', y='Fare', data=train_data, palette='coolwarm')
plt.title('Relación entre la tarifa y la clase de pasajero')
plt.xlabel('Clase de pasajero')
plt.ylabel('Tarifa')
plt.show()
```

```python
#Gráfico de dispersión con ajuste de línea para mostrar la relación entre la edad y la tarifa:
plt.figure(figsize=(10,6))
sns.regplot(x='Age', y='Fare', data=train_data, scatter_kws={"color": "skyblue"}, line_kws={"color": "red"})
plt.title('Relación entre la edad y la tarifa')
plt.xlabel('Edad')
plt.ylabel('Tarifa')
plt.show()
```

```python
#Gráfico de barras apiladas para mostrar la supervivencia según género y clase de pasajero:
plt.figure(figsize=(10,6))
sns.countplot(x='Pclass', hue='Survived', data=train_data, palette='coolwarm')
plt.title('Supervivencia según género y clase de pasajero')
plt.xlabel('Clase de pasajero')
plt.ylabel('Cantidad')
plt.legend(title='Survived', loc='upper right')
plt.show()
```

```python
#Gráfico de densidad para mostrar la distribución de tarifas según la supervivencia:
plt.figure(figsize=(10,6))
sns.kdeplot(x='Fare', data=train_data[train_data['Survived'] == 0], color='skyblue', shade=True)
sns.kdeplot(x='Fare', data=train_data[train_data['Survived'] == 1], color='pink', shade=True)
plt.title('Distribución de tarifas según la supervivencia')
plt.xlabel('Tarifa')
plt.ylabel('Densidad')
plt.legend(['No sobrevivieron', 'Sobrevivieron'], title='Survived', loc='upper right')
plt.show()
```

```python
# Trazar un mapa de calor de correlación entre características
sns.heatmap(train_data.corr(), annot=True, cmap='coolwarm')
```

```python
# Trazar un histograma de distribución por edades
sns.histplot(data=train_data, x='Age', hue='Survived')
```
